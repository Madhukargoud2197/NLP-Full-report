# **NLP PIPELINE**

1: **Data_Acquisition**

Data science team needs to acquire data from the management team which holds records of previous data and stored in any data base or any cloud platform like s3.

In some cases there wont be any data available from the management team, so we need to extracting required info from respective website or any public resource using WEB SCRAPING using Python.



2: **Data_Preprocessing**:

We can discard irrelavant information from the data and change the features based on the requirment.

For example if any ticket is created for a coamplaint, we dont need information about the persons name and time stamp or etc, all we might need is what the severity and whats the text.

So if u have 2 features like title and description, we can combine the both under single feature and use it.

Next correct some mistakes in the data and remove unneeded punctations.


a: **Sentence Segmentation:**


Here we divide the paragraph we have into some n number of sentences, like sentence 1, sentence 2 and so on.

This is also called as sentence tokenization.

We will be using NLTK library and Spacy library to tokenize the sentences.

b: **Word Tokenization:**


Now from the sentences we have, we tokenize the words present in the sentences, so that we can feed them into an ml model and build an NLP application eventually.

c: **Stemming:**

The process of reducing a given word to its base word form or stem is stemming in general terms.

Example:

Hulk loves eating pizza he ate 467 pizzas yesterday


|
   Stemming
|

Hulk *love* *eat*  pizza he ate 467 *pizza* yesterday

d: **Lemmatization:**

The process of coverting a given word with its base word is lemmatization.

Lemmatization understands the word to and reduce it to its base form while in stemming there is no understanding of the meaning of the word, stemming just reduces the word.

for example in the above sentence the word "ate" is the past tense of the word eat. hence after lemmatization, the word ate will be mapped as eat.

Hulk love eat pizza he ate 467 pizza yesterday

{lemmatization}

Hulk love eat pizza he *eat* 467 pizza yesterday.

Difference between **Stemming** and **Lematization**

**Stemming**:

adjustable- adjust

formality- formaliti

formaliti- formal

airliner- airlin



**Lemmatization**:

was-(to) be

better- good

meeting- meeting

This completes our data preprocessing.


3: **Feature Engineering:**

Now our goal is to convert this base words we have into numbers so that a machine can understand it using some methods.

This process of coverting words into numbers and representing them into some vectors can be called as **Feature Engineering.**

Examples: TF-IDF Vectorizer, One Hot encoding, Word Embedding.

4: **Machine Learning Model:**

Now we build an machine learning classification model, not always we might need an ml model, but if your requirment is to classify an tickets into categories, then we build an ML model for sure.

Lets assume we used Naive bayes algorithm and using this data we have prepared, we train the model, then the next time when a ticket is generated at random, this model will categorize the new test data into the respective category.

After building the model we perfom evaluation using confusion matrix. If we feel that the results from evaluation are not good enough, we again start to preprocess the data by making some modifications and the same after that as above process.

Now comes the deployment of the model.

**Model Deployment:**

We can create an fence around the model like using an HTTP link, or using FLASK and deploy the model into the cloud like AZURE or AWS or Google Cloud.

OR

We can deploy it to an end to end machine learning platforms like:

Databricks, AmazonSageMaker, AzureMachinelearning, H20.ai

We make sure that we moniter and keep updating the model if the model is not performing well, hence monitoring is also a very important aspect.

Sometimes we need to customize some words or lemmas as the lemmatization may not understand some words, lets see how we do that.

We use attribute ruler from the pipeline to do that task.


```python
# Lets create a doc
doc = nlp("Bro, do u want to spend some time with your girl? nah bruh, i dont want to have a headache")
for token in doc:
  print(token, '---', token.lemma_)
```

We can see that the algo did not understand bruh and bro for there meaning as brother and gave them as individual words and nah as nah but no, so we can customize these words by using attribute ruler

```python
argm = nlp.get_pipe('attribute_ruler')
argm.add([[{'TEXT':'Bro'}],[{'TEXT':'bruh'}]], {'LEMMA': 'Brother'})
argm.add([[{'TEXT':'nah'}]], {'LEMMA': 'no'})


doc = nlp("Bro, do u want to spend some time with your girl? nah bruh, i dont want to have a headache")
for token in doc:
  print(token, '---', token.lemma_)
```

# **Text Representation**

Feature Engineering is the next most important process that has to be performed in any NLP task.

Feature Engineering is a process of extracting features from a raw data. In NLP, we also call it as Text Representation.

Representing text as a vector is also known as Vector Space Model.


There are various approaches of coverting text into a vector, Namely

1: One Hot Encoding

2:Label Encoding

3: Bag of Words

4: TF-IDF Vectorizer

5: Word Embeddings

Always remember that "Often in NLP, feeding a good text representation to an ordinary algorithm will get you much farther compared to applying a top notch algorithm to an ordinary text representation"



**Label encoding**

Label encoding is basically assigning a number to each word and then creating a list of all the words.


**One Hot encoding**

One Hot encoding, is a technique used to convert categorical data into a binary format where each category is represented by a separate column with a 1 indicating its presence and 0s for all other categories.

Label_onehot snip.png

We ingeneral dont use these in nlp cause of many disadvantages like:

1: Too much memory allocation

2: Cannot understand simialar words in the vector

3: No fixed length representation, which means every time a new word is added into the sentence the size of that vector increases, but we cannot feed diffrent size vector into any ML algo, vector size should be same everytime.

4: When a new word is introduced, which is not in the training set, one hot encoding doesnt recognize the new word, which means this method suffers from Out of vocabulary problem.

```python
nlp = spacy.load("en_core_web_sm")

doc= nlp("Hey Krish, what are you doing to night, can we hang out and watch movies all night?")

for token in doc:
  if token.is_stop:
    print(token)
```

```python
from operator import is_not
for token in doc:
  if token.is_stop == False:
    print(token)
```

```python
# Lets write a function so that we can always use that fucntion to remove stopwords and get a sentence without stopwords in it.
def preprocess(text):
  doc= nlp(text)

  Final_array= [token.text for token in doc if token.is_stop == False]
  return " ".join(Final_array)
```

# **TF-IDF**

Term frequency-inverse document frequency (TF-IDF) is a technique used in natural language processing (NLP) to measure how important a word is in a document.

As we may encounter many words that are similar in all the different documents, we need to find a way to lower there impact,
So if the word is more often repeated in all the documents, it should have less impact value, than the word that is less repeated in the documents.

This can be done by dividing the total number of documents by the number of times that particular word t appeared.

IDF= (Total documents/ No of documents with term t in it).

for decresing the impact of this on the processing we apply log transformation.

IDF= log(Total documents/ No of documents with term t in it).



TF_IDF.png

Now we also need to check how much that word has its impact on that particular document, so that we can understand its own impact in the document itself, for that we calculate term frequency.

TF(t,d)= (Total no of times t present in doc x/ total no of words in the doc x)

Hence we can say that

TF- IDF = TF(t,d)* IDF(t)

```python
# Lets perform this in code
# lets write some sentences an assume them to be documents

Collection_ = [" Thor eating pizza, Loki is eating pizza, Ironman ate pizza already",
               " Apple is announcing new iphone tomorrow",
               "Tesla is announcing new model-3 tomorrow",
               "Google is annoncing new pixel phone 5 tomorrow",
               "Amazon is annoncing new eco-dot tomorrow",
               " I am eating biryani and are you eating grapes",
               " Something is amazing"]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer
vector_TF= TfidfVectorizer()
Transformed_vector= vector_TF.fit_transform(Collection_)
print(vector_TF.vocabulary_)
```

```python
feature_names= vector_TF.get_feature_names_out()
feature_names
```

# **Gensim**

Gensim is a free, open-source Python library used for unsupervised topic modeling and Natural Language Processing (NLP) tasks. It's designed to extract semantic concepts from documents, build vector representations of words or documents, and perform various NLP tasks using modern statistical machine learning algorithms.




##Key features and capabilities:



**Topic Modeling:**

 Gensim implements algorithms like Latent Dirichlet Allocation (LDA) and Latent Semantic Indexing (LSI) to discover underlying topics in a collection of documents.


**Vector Representation:**


It provides tools for creating word or document vectors, allowing for semantic comparisons and similarity analysis.


**Large Corpus Handling:**


Gensim is designed to handle large text collections efficiently, utilizing data streaming and online algorithms.


**Various NLP Tasks:**


It supports tasks such as document indexing, retrieval by similarity, and other NLP functionalities.


**Popular Algorithms:**


Gensim incorporates widely used algorithms like Word2Vec and FastText for creating word embeddings and understanding word relationships.


**Performance and Optimization:**


Gensim is implemented in Python and Cython for performance, and it leverages Python's built-in generators and iterators for efficient data processing
